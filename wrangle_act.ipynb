```python
# referenced from knowledge
tweet_data = [] #creating an empty list to store the data from json.txt file
with open("tweet-json.txt", "r") as json_file: #opening 'tweet-json.txt' file using with statement
    for line in json_file: #iterating over lines in the file
        json_tweet_data = json.loads(line) #parsing json file data into lines
        
        tweet_id = json_tweet_data['id'] #extracting the tweet id in a variable
        favorites = json_tweet_data['favorite_count'] #extracting the tweet favotire count in a variable
        retweets = json_tweet_data['retweet_count'] #extracting the tweet retweet count in a variable
        
        #appending the above data to tweet_data empty list
        tweet_data.append({"tweet_id":tweet_id, "favorites":favorites, "retweets":retweets})
        
td_df = pd.DataFrame(tweet_data)  #creating a dataframe for tweet_data list 
```
